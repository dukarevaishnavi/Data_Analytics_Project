```python
# Which TV Show has the highest number of seasons?
tv = netflix[netflix["type"] == "TV Show"]

tv["duration"] = tv["duration"].str.replace(" Seasons", "").str.replace(" Season", "").astype(int)

tv.sort_values(by="duration", ascending=False)[["title", "duration"]].head(1)
```

```python
# Director & Cast Analysis
```

```python
# Which director has directed the most content on Netflix?
netflix["director"].value_counts().head()
```

```python
# Which actor appears most frequently?
netflix.columns
netflix["cast"].dropna().str.split(",").explode().value_counts().head(1)
```

```python
# Which actors appear in the highest number of movies?
netflix[netflix["type"]=="Movie"]["cast"].str.split(", ").explode().value_counts().head(1)
```

```python
# Which directors worked on both Movies and TV Shows?
set(netflix[netflix["type"]=="Movie"]["director"].dropna()) & set(netflix[netflix["type"]=="TV Show"]["director"].dropna())
```
